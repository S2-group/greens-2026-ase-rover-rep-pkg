## Here, will be covered the measurements done using powerstat and the energy service + analyser service

```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import re
import numpy as np
```

## Please modify root_path in order to switch experiments (idle or imaging)

```python
# followed by folders each with their own .txt file
root_path = Path("./data/measurements_idle_state/experiments/new_runner_experiment/")
dir_list = os.listdir(root_path)

# Filter + sort by repetition number
sorted_dirs = sorted(
    [item for item in dir_list if re.match(r"run_0_repetition_(\d+)", item)],
    key=lambda name: int(re.search(r"run_0_repetition_(\d+)", name).group(1))
)

# Print sorted
print(sorted_dirs)
```

```python
# data = pd.read_csv('./data/measurements_idle_state/experiments/new_runner_experiment/run_0_repetition_0/analyser_logs_clean.txt', sep=',')
# # print(data)
```

## Get path for experiment and order it 

```python
all_logs = []

for folder_name in sorted_dirs:
    folder_path = root_path / folder_name
    txt_path = folder_path / "analyser_logs_clean.txt"

    if not txt_path.exists():
        continue  # skip if file doesn't exist
        
    with open(txt_path, "r") as f:
        lines = f.readlines()

        clean_lines = []
        
        for line in lines:
            match = re.search(r"\[(\d{2}:\d{2}:\d{2})\]\s*(EnergySensorOutput.*?)$", line)
            if match:
                timestamp = match.group(1)
                log_data = match.group(2)
                clean_lines.append({
                    "time": timestamp,
                    "data": log_data
                })
        all_logs.append({
            "PATH": txt_path,
            "lines": clean_lines
        })

# print(all_logs)
print(all_logs[4]["PATH"])
all_logs[4]['lines']
```

## Convert to dataframe

```python
records = []

for log in all_logs:
    for entry in log["lines"]:
        records.append({
            "file": log["PATH"].parent.name,
            "timestamp": entry["time"],
            "data": entry["data"]
        })
```

```python
df = pd.DataFrame(records)
display(df)
```

## Format data into DataFrame, along with voltage, current and power => clean

```python
def extract_values(data):
    match = re.search(r"supply_voltage=(.*?), current_amps=(.*?), power_watts=(.*?)\)", data)
    if match:
        return float(match.group(1)), float(match.group(2)), float(match.group(3))
    return None, None, None

df[["voltage", "current", "power"]] = df["data"].apply(
    lambda x: pd.Series(extract_values(x))
)
df.drop('data', axis=1, inplace=True)
display(df)
```

## Plot based on each experiment run (30 repetitions per run)

```python
# STEPS:
# get first col of df
# filter such that the last one or 2 digits of the name represent each run 
# for each filtered df, compute plot for the 3 variables (voltage, current, power) based on timestamp
# types of plots : 
# 1. line chart with a line for each variable plotted against the timestamp
# 2. scatter plot with a color for each variable plotted against the timestamp
```

## Shapiro-Wilk test for normality. If p > 0.05, the data is normal.
In our test below (for the imaging state), this appears to be normal for 28 repetitions, and false only for 2. Based on research, it seems like a good result for energy measurements.

```python
from scipy.stats import shapiro

for rep in df["repetition"].unique():
    p = shapiro(df[df["repetition"] == rep]["power"])[1]
    print(f"Repetition {rep} normal? {p > 0.05}")
```

## Compute average measurements for all logs for each repetition and plot a line chart 
Each repetition 0,1,2 .. has multiple measurements. Here, I compute the average of the measurements per repetition and I plot based on those averages

```python
# Group by repetition and compute mean for voltage, current, and power
avg_per_run = df.groupby("repetition")[["voltage", "current", "power"]].mean().reset_index()
avg_per_run.head()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(avg_per_run["repetition"], avg_per_run["voltage"], label="Voltage (V)")
plt.plot(avg_per_run["repetition"], avg_per_run["current"], label="Current (A)")
plt.plot(avg_per_run["repetition"], avg_per_run["power"], label="Power (W)")

plt.title("Average Voltage, Current, and Power per Run")
plt.xlabel("Repetition")
plt.ylabel("Average Measurement")
plt.xticks(avg_per_run["repetition"])  # Show all run numbers
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Create X locations
reps = avg_per_run["repetition"].values
space = 3
x = np.arange(len(reps)) * space  # positions on x-axis

# 3. Set bar width
width = 0.6

# 4. Plot bars for each metric
plt.figure(figsize=(16, 8))
plt.bar(x - width, avg_per_run["voltage"], width, label="Voltage (V)", color="tab:blue")
plt.bar(x,         avg_per_run["current"], width, label="Current (A)", color="tab:orange")
plt.bar(x + width, avg_per_run["power"],   width, label="Power (W)",   color="tab:green")

# 5. Add labels and formatting
plt.xlabel("Repetition")
plt.ylabel("Average Value")
plt.title("Average Voltage, Current, and Power per Repetition")
plt.xticks(x, reps)
plt.legend()
# plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
